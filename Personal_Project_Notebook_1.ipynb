```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-



import requests
from bs4 import BeautifulSoup
import pandas as pd


class TFFRS:
    def __init__(self,conf_number,Event):
        self.Conference=conf_number
        self.Event=Event
    def BuildDataFrame(self):




        # Initialize an empty DataFrame
        all_data_df = pd.DataFrame(columns=['Name', 'School', 'Jump (m)', 'Conference', 'Year'])

        conf_number=self.Conference
        Event=self.Event


        if Event=='Dec':
            url = f"https://www.tfrrs.org/printable_lists/{conf_number}?other_lists=https%3A%2F%2Ftf.tfrrs.org%2Flists%2F4278%2FAtlantic_10_Outdoor_Performance_List&limit=30&event_type=&year=&gender=&performance_event_hnd=31632"
        if Event=='Jav':
            url = f"https://www.tfrrs.org/printable_lists/{conf_number}?other_lists=https%3A%2F%2Ftf.tfrrs.org%2Flists%2F4278%2FAtlantic_10_Outdoor_Performance_List&limit=30&event_type=&year=&gender=&performance_event_hnd=31629"
        if Event=='Hammer':
            url = f"https://www.tfrrs.org/printable_lists/{conf_number}?other_lists=https%3A%2F%2Ftf.tfrrs.org%2Flists%2F4278%2FAtlantic_10_Outdoor_Performance_List&limit=30&event_type=&year=&gender=&performance_event_hnd=31627"
        if Event=='Discus':
            url = f"https://www.tfrrs.org/printable_lists/{conf_number}?other_lists=https%3A%2F%2Ftf.tfrrs.org%2Flists%2F4278%2FAtlantic_10_Outdoor_Performance_List&limit=30&event_type=&year=&gender=&performance_event_hnd=31626"
        if Event=='Shotput':
            url = f"https://www.tfrrs.org/printable_lists/{conf_number}?other_lists=https%3A%2F%2Ftf.tfrrs.org%2Flists%2F4278%2FAtlantic_10_Outdoor_Performance_List&limit=30&event_type=&year=&gender=&performance_event_hnd=31623"
        if Event=='High':
            url = f"https://www.tfrrs.org/printable_lists/{conf_number}?other_lists=https%3A%2F%2Ftf.tfrrs.org%2Flists%2F4278%2FAtlantic_10_Outdoor_Performance_List&limit=30&event_type=&year=&gender=&performance_event_hnd=31615"
        if Event=='Long':
            url = f"https://www.tfrrs.org/printable_lists/{conf_number}?other_lists=https%3A%2F%2Ftf.tfrrs.org%2Flists%2F4278%2FAtlantic_10_Outdoor_Performance_List&limit=30&event_type=&year=&gender=&performance_event_hnd=31620"
        if Event=='Pole Vault':
            url = f"https://www.tfrrs.org/printable_lists/{conf_number}?other_lists=https%3A%2F%2Ftf.tfrrs.org%2Flists%2F4278%2FAtlantic_10_Outdoor_Performance_List&limit=30&event_type=&year=&gender=&performance_event_hnd=31617"
        if Event=='Triple':
            url = f"https://www.tfrrs.org/printable_lists/{conf_number}?other_lists=https%3A%2F%2Ftf.tfrrs.org%2Flists%2F4278%2FAtlantic_10_Outdoor_Performance_List&limit=30&event_type=&year=&gender=&performance_event_hnd=31621"

        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')

        # Initialize lists to store data
        names = []
        schools = []
        jumps = []
        conferences = []
        years = []

        # Locate and store names, schools, jumps, and years
        for row in soup.find_all('tr', class_='allRows'):
            cols = row.find_all('td')

            if len(cols) >= 5:
                names.append(cols[1].text.strip())
                years.append(cols[2].text.strip())  # added line to scrape "Year" info
                schools.append(cols[3].text.strip())
                jumps.append(float(cols[4].text.strip()[:-1]))
                conferences.append(self.Conference)  # Conference number is appended here

        # Create a DataFrame
        df = pd.DataFrame({
            'Name': names,
            'School': schools,
            'Jump (m)': jumps,
            'Conference': conferences,
            'Year': years  # added 'Year' to the DataFrame
        })

        # Append the DataFrame
        all_data_df = pd.concat([all_data_df, df])

            # Pause for a bit before the next request
        return all_data_df


'''
test=TFFRS(4277,'Triple')
test.BuildDataFrame()
testdf=test.BuildDataFrame()
'''



```

```python
import pandas as pd

def merge_csv_times(csv1_path, csv2_path, output_path):
    # Read CSV files into pandas DataFrames
    df1 = pd.read_csv(csv1_path)
    df2 = pd.read_csv(csv2_path)

    # Remove the 'Meet' and 'Meet Date' columns from both DataFrames
    df1 = df1.drop(columns=['Meet', 'Meet Date'])
    df2 = df2.drop(columns=['Meet', 'Meet Date'])

    # Merge the two DataFrames based on 'Athlete' and 'Team'
    merged_df = pd.merge(df1, df2, on=['Athlete', 'Team'], suffixes=('', '_next_year'))

    # Rename the 'Time_next_year' column to 'Next Year Time'
    merged_df.rename(columns={'Time_next_year': 'Next Year Time'}, inplace=True)

    # Save the merged DataFrame to a new CSV file
    merged_df.to_csv(output_path, index=False)

# Replace 'path_to_csv1.csv' and 'path_to_csv2.csv' with the actual file paths of your CSVs
csv1_path = "D1 - Men's 100m 2022.csv"
csv2_path = "D1 - Men's 100m 2023.csv"
output_path = 'merged_times.csv'
#path 1 should be the earlier year and path 2 should be the more recent year

merge_csv_times(csv1_path, csv2_path, output_path)
```

```python
def year_to_numeric(year_code):
    if year_code.startswith('JR'):
        return int(year_code.split('-')[1])
    elif year_code.startswith('SR'):
        return int(year_code.split('-')[1])
    elif year_code.startswith('FR'):
        return int(year_code.split('-')[1])
    elif year_code.startswith('SO'):
        return int(year_code.split('-')[1])
    else:
        return None

# Apply the function to the 'Year' column
m100mxtrain['Year'] = m100mxtrain['Year'].apply(year_to_numeric)
```
